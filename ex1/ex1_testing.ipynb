```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.binomial(1, 0.25, (100000, 1000))
epsilon = 0.5
```

```python
cummean = lambda x:  x.cumsum()/np.arange(1, len(x)+1)
means = np.abs(np.apply_along_axis(cummean, 1, data)-0.25)
```

```python
pct = lambda x: np.where(x>=epsilon)[0].shape[0] / 100000
```

```python
np.apply_along_axis(pct, 1,means)
```
